```python
import pandas as pd

df = pd.read_csv("./data/bgg_dataset.txt", sep=";")

df.head()

print(len(df))  # Get the number of rows in the DataFrame

for col in df.columns:
    print(f"{col}: {df[col].isna().sum()}")

# print the first 10 rows numbers wehre there are empty cells in domain column

for i in range(len(df)):
    if df.loc[i, "Domains"] == "":
        print(i)


```

```python
# print the first 10 rows numbers wehre there are empty cells in domain column
count = 0
for i in range(len(df)-1, -1, -1):
    if pd.isna(df.loc[i, "Domains"]):
        print(i)
        count += 1
        if count >= 100:  # Stop after printing 10 rows
            break
```
